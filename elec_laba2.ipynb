```python
from pandas import *
from numpy import *
from bokeh.plotting import figure, output_file, show
from bokeh.io import output_notebook

output_notebook()
csv_data = read_csv("https://drive.google.com/uc?export=download&id=1VRICeFtVx9dwuTSIDTlVSQzRqQjQVHZU") 
##у данной таблицы нет оглавления, она состоит из одного столбца со значениями
input_matrix = csv_data.values
p = figure(plot_width = 1600, plot_height = 400)

p.line(arange(len(input_matrix)), input_matrix[:,0], line_width = 2)
a = input_matrix[:,0]
for i in range(len(a)):
  if input_matrix[i] > 35:
    p.dot(i, a[i], color = 'green', size = 20)
  else:
    p.dot(i, a[i], color = 'red', size = 20)
##зеленые точки-значения, при которых будет выключаться освещение
##красные- при которых оно включается
p.ray(x = 0.0, y=35.0, line_color = 'orange', line_width = 2)
##
csv_data.head()
csv_data.tail()
show(p)
```
